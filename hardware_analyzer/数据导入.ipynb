```python
import os
import csv
import django
from django.conf import settings

# 设置 Django 环境
os.environ.setdefault('DJALO_SETTINGS_MODULE', 'hardware_analyzer.settings')
os.environ['DJANGO_ALLOW_ASYNC_UNSAFE'] = 'true'
django.setup()

from analyzer.models import PowerSupply

def import_power_supply_data(csv_file_path):

    PowerSupply.objects.all().delete()
    # 导入 CSV 数据
    with open(csv_file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            PowerSupply.objects.create(
                title=row['标题'],
                reference_price=row['参考价'] or None,
                jd_price=row['京东价'] or None,
                jd_link=row['京东链接'] or None,
                product_image=row['产品图片'] or 'https://example.com/default.jpg',
                product_parameters=row['产品参数'] or None
            )

    # 检查导入结果
    print(f"成功导入 {PowerSupply.objects.count()} 条电源数据")
    for power_supply in PowerSupply.objects.all():
        print(f"ID: {power_supply.id}, Title: {power_supply.title}, 参考价: {power_supply.reference_price}, 京东价: {power_supply.jd_price}, 京东链接: {power_supply.jd_link}, 产品图片: {power_supply.product_image}, 产品参数: {power_supply.product_parameters}")

if __name__ == "__main__":
    csv_file_path = '../spider/csv/电源01.csv'  # 默认路径，可改为实际路径
    import_power_supply_data(csv_file_path)
```

```python
import os
import csv
import django
from django.conf import settings

# 设置 Django 环境
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'hardware_analyzer.settings')
os.environ['DJANGO_ALLOW_ASYNC_UNSAFE'] = 'true'
django.setup()

from analyzer.models import Cooler

def import_cooler_data(csv_file_path):
    # 导入 CSV 数据
    with open(csv_file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            Cooler.objects.create(
                title=row['标题'],
                reference_price=row['参考价'] or None,
                jd_price=row['京东价'] or None,
                jd_link=row['京东链接'] or None,
                product_image=row['产品图片'] or 'https://example.com/default.jpg',
                product_parameters=row['产品参数'] or None
            )

    # 检查导入结果
    print(f"成功导入 {Cooler.objects.count()} 条散热器数据")
    for cooler in Cooler.objects.all():
        print(f"ID: {cooler.id}, Title: {cooler.title}, 参考价: {cooler.reference_price}, 京东价: {cooler.jd_price}, 京东链接: {cooler.jd_link}, 产品图片: {cooler.product_image}, 产品参数: {cooler.product_parameters}")

if __name__ == "__main__":
    csv_file_path = '../spider/csv/散热01.csv'  # 默认路径，可改为实际路径
    import_cooler_data(csv_file_path)
```
